## 3. Цели и план проекта

### 3.1 Что реализуем и что проверяем?

В рамках проекта мы воспроизводим подходы из статьи Neural Stochastic Differential Equations и проверяем их эффективность к задаче прогнозирования финансовых временных рядов

#### 1. Репликация
- Реализуем базовую модель neural SDE с нейросетевыми drift и diffusion
- Реализация ELBO (Evidence Lower BOund)
- Обучение на синтетических данных (как в статье)
- Воспроизведение результатов из статьи

#### 2. Применение к финансовым данным
- Проноз цены акций на данных Московской бирже
- Сбор данных и предобработка (временные ряды цен / логарифм доходностей)
- Обучение модели на реальных данных

#### 3. Сравнение с альтернативами
- Сравнение модели с:
    - Neural ODE
    - классическими моделями ГО
    - классическими финансовыми моделями прогнозирования динамики цен

#### 4. Визуализация
- Подготовка UI-визуализации прогнозирования цен

### 3.2 Метрики успеха

#### 1. Репликация
- Корректная реализация ELBO
- Сходимость обучения
- Визуальное совпадение с результатами статьи

#### 2. Качество модели (на данных цен финансовых активов)
- MSE / MAE для цен или доходностей
- Лог-правдоподобие
- Сравнение реальных и сгенерированных траекторий

#### 3. Сравнение моделей
- Сравнение Neural SDE vs Neural ODE:
    - точность
    - устойчивость
    - способность моделировать стохастичность
- Сравнение Neural SDE и классических финансовых моделей
    - точность
    - MSE

#### 4. Визуализация
- построение реальных vs сгенерированных траекторий
- построение доверительных интервалов прогноза

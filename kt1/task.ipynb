##[**Neural Stochastic Differential Equations: Deep Latent Gaussian Models in the Diffusion Limit**](https://arxiv.org/pdf/1905.09883)

В статье предлагается рассматривать глубокие вероятностные модели в пределе бесконечного числа слоёв, что приводит к непрерывной стохастической динамике.

Основная идея состоит в том, чтобы заменить дискретную генерацию латентных переменных на непрерывный стохастический процесс, задаваемый стохастическим дифференциальным уравнением. Это позволяет формулировать модель непосредственно на пространстве траекторий.

## 1. Постановка задачи

Мы хотим построить генеративную модель, в которой:

- латентная переменная — это траектория стохастического процесса
- модель задаётся стохастическим дифференциальным уравнением
- нейросеть параметризует функции дрейфа и диффузии

Формально, латентный процесс задаётся как:

$$
dX_t = b(X_t, t)\,dt + \sigma(X_t, t)\,dW_t
$$

где:
- $b(x,t)$ — drift
- $\sigma(x,t)$ — diffusion
- $W_t$ — броуновское движение

Наблюдение:
$$
Y \sim p(y \mid X_1)
$$

## 1.1 Мотивация: зачем SDE вместо ODE?

### Связь: Neural ODE → Neural SDE

Neural ODE описывает **детерминированную динамику**:
$$
dX_t = f(X_t, t)\,dt
$$

Однако в генеративных моделях важна **стохастичность**, так как:
- данные содержат шум
- требуется моделировать распределения, а не одну траекторию

### Почему SDE лучше:

| Свойство | ODE | SDE |
|--------|-----|-----|
| Стохастичность | нет | есть |
| Генерация разнообразия | отсутствует | присутствует |
| Моделирование неопределённости | ограничено | выражено явно |
| Работа с вероятностными распределениями | ограничена | корректно формализована |

### Ключевая идея

Вместо детерминированного потока распределений рассматриваются случайные траектории процесса, что приводит к стохастической формулировке динамики

### Связь ODE и SDE

Стохастическое дифференциальное уравнение можно рассматривать как обобщение детерминированного случая. Если диффузионный член отсутствует, уравнение переходит в ODE.

В модели Neural SDE динамика задаётся уравнением:

$$
dX_t = b(X_t, t)\,dt + \sigma(X_t, t)\,dW_t
$$

где второй член отвечает за случайную составляющую. Наличие этого члена приводит к тому, что каждому начальному состоянию соответствует не одна траектория, а распределение траекторий

### Переход от ODE к SDE

В Deep Latent Gaussian Models (DLGM):

$$
X_i = X_{i-1} + b_i(X_{i-1}) + \sigma_i Z_i
$$

Если:
- число слоёв $\to \infty$
- шаг $\to 0$
- шум $\to 0$

$\to$ получаем SDE:

$$
dX_t = b(X_t, t)\,dt + \sigma(X_t, t)\,dW_t
$$

Это и есть **диффузионный предел DLGM**

## 1.2 Формальная постановка задачи

В качестве отправной точки рассматриваются Deep Latent Gaussian Models. В этих моделях латентная переменная строится как последовательность состояний, получаемых из предыдущего состояния с добавлением гауссовского шума.

Формально процесс задаётся рекурсией:

$$
X_0 = Z_0
$$

$$
X_i = X_{i-1} + b_i(X_{i-1}) + \sigma_i Z_i
$$

где $Z_i \sim \mathcal{N}(0, I)$. Наблюдаемая переменная генерируется по последнему состоянию цепи.

Основной задачей является вычисление маргинального правдоподобия, которое требует интегрирования по всем латентным переменным и, как правило, не имеет аналитического решения

### Variational Inference (ELBO)

Для решения задачи используется вариационный принцип. Логарифм маргинального правдоподобия представляется в виде оптимизационной задачи:

$$
\log p_\theta(y) \geq \mathbb{E}_{q_\beta(z \mid y)}[\log p_\theta(y \mid z)] - D_{KL}(q_\beta(z \mid y) \parallel p(z))
$$

Это даёт:
- обучаемую цель (ELBO)
- возможность использовать backpropagation

## 1.3 Диффузионный предел

Ключевой шаг состоит в переходе к непрерывному пределу. Рассматривается ситуация, в которой число слоёв стремится к бесконечности, а шаг дискретизации и уровень шума уменьшаются.

В этом пределе латентная переменная описывается стохастическим дифференциальным уравнением:

$$
dX_t = b(X_t, t)\,dt + \sigma(X_t, t)\,dW_t
$$

Таким образом, дискретная модель переходит в непрерывную, а латентное пространство становится пространством непрерывных функций.

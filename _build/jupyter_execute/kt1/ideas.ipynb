## 2. Ключевые идеи статьи

После перехода к непрерывному времени вся случайность модели сосредоточена в броуновском движении. Решение стохастического уравнения можно рассматривать как отображение, которое каждой реализации броуновского движения сопоставляет траекторию процесса.

Это является непрерывным аналогом представления дискретной модели через базовые гауссовские шумы и позволяет формулировать генеративную модель как преобразование на пространстве траекторий.

### 2.1 Параметризация через нейросеть

В SDE:

$$
dX_t = b(X_t, t; \theta)\,dt + \sigma(X_t, t; \theta)\,dW_t
$$

где:
- $b$ — нейросеть (drift)
- $\sigma$ — нейросеть (diffusion)

Функции дрейфа и диффузии в стохастическом уравнении задаются параметрическими отображениями, реализуемыми нейросетями. При фиксированных параметрах решение SDE определяет отображение из пространства броуновских траекторий в пространство латентных переменных.

Наблюдаемая переменная зависит от конечного состояния процесса, что полностью задаёт вероятностную модель.

### 2.2 Теорема Гирсанова и вариационный вывод

В непрерывной постановке вариационный вывод переносится на пространство траекторий. Существенную роль играет теорема Гирсанова, которая устанавливает связь между изменением меры и добавлением дрейфа к броуновскому движению.

Любое вариационное распределение может быть представлено в виде процесса:

$$
Z_t = W_t + \int_0^t u_s ds
$$

При этом KL-дивергенция выражается через интеграл квадрата нормы функции. Это позволяет получить вариационную функциональную, аналогичную ELBO, но определённую для непрерывного времени

### 2.3 Стохастическое автодифференцирование

Проблема: нельзя напрямую применить adjoint method (как в ODE)

Причина:
- зависимость от всей траектории
- forward-backward SDE сложны

### Решения:

1. **Discretize → Backprop (Euler)**
   - аналог DLGM
   - простой, но дискретизация


2. **Pathwise differentiation**
   - дифференцируем SDE
   - используем теорию стохастических потоков
